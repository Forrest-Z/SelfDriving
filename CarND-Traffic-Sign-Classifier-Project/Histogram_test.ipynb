```python

import matplotlib.pyplot as plt
# Visualizations will be shown in the notebook.
%matplotlib inline
from numpy.random import normal, uniform
gaussian_numbers = normal(size=1000)
# plt.hist(gaussian_numbers, bins=30, normed=True, cumulative=True)
# plt.hist(gaussian_numbers, bins=(-10, -1, 1, 10))
# plt.hist(gaussian_numbers, bins=20, histtype='step')
uniform_numbers = uniform(low=-3, high=3, size=1000)
plt.hist(gaussian_numbers, bins = 30, normed = True, histtype='stepfilled', color='b', label='Gaussian')
plt.hist(uniform_numbers, bins = 30, normed = True, histtype = 'stepfilled', color='r', alpha=0.5, label='Uniform')
plt.title('Gaussian Histogram')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
import pylab as P

# true line shape with equally spaced histogram plot

# create an array of numbers belonging to gaussian distribution 
mu,sigma = 200, 25
x = mu + sigma * P.randn(10000)

# the histogram of the data with histtype = 'step'
n, bins, patches = P.hist(x, bins = 50, normed = True, histtype='stepfilled')
P.setp(patches, 'facecolor', 'y', 'alpha', 0.85)

# add a line showing the expected distribution
y = P.normpdf(bins, mu, sigma)
l = P.plot(bins, y, 'g--', linewidth=2.5)
```

```python
# uneaully spaced bin edges

# create a histogram by providing the bin edges (unequally spaced)
P.figure()

bins = [100, 125, 150, 160, 170, 180, 190, 200, 210, 220, 230, 240, 250, 275, 300]

# the histogram of the data with histtype = 'step'
n, bins, patches = P.hist(x, bins, normed=True, histtype='bar', rwidth = 0.8)
```

```python
# now a cumulative histogram of the data
P.figure()

n, bins, patches = P.hist(x, bins = 50, normed=True, histtype='step', cumulative = True)

# adding a line showing the expected distribution
y = P.normpdf(bins, mu, sigma).cumsum()
y /=y[-1]
l = P.plot(bins, y, 'k--', linewidth = 1.5)

# create a second dataset with smaller standard deviation 
sigma2 = 15
x2 = mu + sigma2 * normal(size = 10000)

n, bins, patches = P.hist(x2, bins=50, normed = True, histtype ='step', cumulative= True)
y2 = P.normpdf(bins, mu, sigma2).cumsum()
y2 /= y2[-1]
l2 = P.plot(bins, y2, 'g--', linewidth = 1.5)

# finally overplot a reverted cumulative histgram
n, bins, patches = P.hist(x2, bins=bins, normed=True, histtype='step', cumulative = -1)

P.grid(True)
P.ylim(0, 1.05)
```

```python
# histogram has the ability to plot multiple data in parallel ...
# Note the new color kwarg, used to override the default, which uses the line color cycle

P.figure()
x = mu + sigma * P.randn(1000, 3)

n, bins, patches = P.hist(x, bins=10, normed=1, histtype = 'bar', 
                         color = ['crimson', 'burlywood', 'chartreuse'],
                         label = ['crimson', 'burlywood', 'chartreuse'])
P.legend()
```

```python
# stack using the step histtype

P.figure()
n, bins, patches = P.hist(x, bins= 10, normed =1, histtype='bar', rwidth = 0.7, stacked = True)
P.show()

print('bins={}'.format(bins))

P.figure()
n, bins, patches = P.hist(x, 10, normed =1, histtype='step', stacked = True)
P.show()

```
